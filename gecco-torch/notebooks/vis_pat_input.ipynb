```python
from gecco_torch.scatter import plot, combine_plots
import torch
```

```python
pth = "/nrs/turaga/jakob/autoproof_data/flywire_cave_post/debug/"
idx = 5
gen = torch.load(pth + f"generated_{idx}.pt")
partial = torch.load(pth + f"partial_{idx}.pt")
superset = torch.load(pth + f"superset_{idx}.pt")
gt = torch.load(pth + f"gt_{idx}.pt").squeeze(0)

print(gt.shape)

# randomly select 10_000 points from superset
if superset.shape[0] > 10_000:
    indices = torch.randperm(superset.shape[0])[:10_000]
    superset = superset[indices]

print(partial.shape, gen.shape, superset.shape)
```

```python
combined = torch.cat([gen, gt], dim=0)  # combine the partial and generated point clouds
combined_labels = torch.cat([
    torch.ones(gen.shape[0]) * 0,  # label for generated points
    torch.ones(gt.shape[0]) * 1  # label for ground truth points
], dim=0)

plot(combined.cpu().numpy(), combined_labels.cpu().numpy(), label_map={0: 'Generated', 1: 'Ground Truth'})
```

```python
all_pts = torch.cat([superset, gt], dim=0)
all_labels = torch.cat([torch.zeros_like(superset), torch.ones_like(gt)], dim=0)[..., 0]

plot(all_pts, all_labels, label_map= {0: 'Superset', 1: 'Ground Truth'}, size=1.5)
```
