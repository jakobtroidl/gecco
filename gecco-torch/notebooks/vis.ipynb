```python
from gecco_torch.data.neurons import TorchNeuronNet
from gecco_torch.scatter import plot
import torch
```

```python
root = "/nrs/turaga/jakob/autoproof_data/flywire_cave_post/cache_t5"
split = "train"
n_points = 2048
```

```python
data = TorchNeuronNet(root, split, n_points)
```

```python
pts = data[0].points
partial = data[0].partial

combined = torch.cat([pts, partial], dim=0)
combined_labels = torch.cat([torch.zeros(pts.shape[0]), torch.ones(partial.shape[0])], dim=0)
```

```python
plot(combined.cpu().numpy(), labels=combined_labels.cpu().numpy())
```
